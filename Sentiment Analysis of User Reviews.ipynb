```python
# Import necessary libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import pandas as pd

# Ensure necessary NLTK data files are downloaded
nltk.download('punkt')
nltk.download('wordnet')

# Sample data (Assuming 'data' is your DataFrame with a 'Translated_Review' column)
data = pd.DataFrame({
    'Translated_Review': [
        "The app is great! Highly recommend it.",
        "Poor service, not worth the money.",
        "I love this app, it's really helpful.",
        "The app crashes frequently, terrible experience."
    ]
})

# Step 1: Tokenization
def tokenize_text(text):
    return word_tokenize(text)

# Apply tokenization to the 'Translated_Review' column
data['Tokens'] = data['Translated_Review'].apply(tokenize_text)

# Step 2: Stemming and Lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def stem_and_lemmatize(tokens):
    # Stemming
    stemmed = [stemmer.stem(token) for token in tokens]
    # Lemmatization
    lemmatized = [lemmatizer.lemmatize(token) for token in stemmed]
    return lemmatized

# Apply stemming and lemmatization to the tokenized data
data['Processed_Tokens'] = data['Tokens'].apply(stem_and_lemmatize)

# Display the original tokens and processed tokens (stemmed and lemmatized)
print(data[['Tokens', 'Processed_Tokens']].head())
```

```python
# 2- Handling Negations and Sarcasm
# A. Handling Negations
def handle_negation(tokens):
    new_tokens = []
    negation = False
    
    for token in tokens:
        if token in ["not", "no", "never"]:  # Add more negation words as needed
            negation = True
        elif negation:
            new_tokens.append("not_" + token)  # Prefix "not_" to the following token
            negation = False
        else:
            new_tokens.append(token)
    
    return new_tokens

# Apply the negation handling function
data['Negation_Handled_Tokens'] = data['Processed_Tokens'].apply(handle_negation)
print(data[['Processed_Tokens', 'Negation_Handled_Tokens']].head())
```

# Time Series Graph for Sentiment Trends Over Time

```python
import pandas as pd
import numpy as np

# Sample data with dates and sentiment
date_rng = pd.date_range(start='2024-01-01', end='2024-01-10', freq='D')
sentiment_values = np.random.choice(['Positive', 'Negative', 'Neutral'], size=len(date_rng))

# Creating a DataFrame
time_series_data = pd.DataFrame(date_rng, columns=['Date'])
time_series_data['Sentiment'] = sentiment_values

# Grouping by date and sentiment for counting
sentiment_counts = time_series_data.groupby(['Date', 'Sentiment']).size().unstack(fill_value=0)

# Plotting the time series graph
plt.figure(figsize=(12, 6))
sentiment_counts.plot(kind='line', marker='o')
plt.title('Sentiment Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid()
plt.legend(title='Sentiment')
plt.show()
```

# Visualizing Text Data Cleaning Techniques

```python
# Import necessary libraries
from sklearn.feature_extraction.text import CountVectorizer
import string
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Text Cleaning Function
def clean_text(reviews):
    """Cleans the reviews by lowercasing and removing punctuation."""
    cleaned_reviews = []
    for review in reviews:
        # Convert to lowercase
        review = review.lower()
        # Remove punctuation
        review = review.translate(str.maketrans("", "", string.punctuation))
        cleaned_reviews.append(review)
    return cleaned_reviews

# Step 2: Prepare the Data
# Assuming df is a DataFrame containing the 'Review' column
original_reviews = df['Review'].dropna().tolist()  # Get list of reviews

# Step 3: Clean the Reviews
cleaned_reviews = clean_text(original_reviews)

# Step 4: Vectorization (Word Frequency Count)
# Create two CountVectorizers: one for original and one for cleaned data
vectorizer_original = CountVectorizer()
vectorizer_cleaned = CountVectorizer()

# Fit and transform both original and cleaned reviews
X_original = vectorizer_original.fit_transform(original_reviews)
X_cleaned = vectorizer_cleaned.fit_transform(cleaned_reviews)

# Step 5: Calculate Word Frequencies
# Sum the occurrences of each word in original and cleaned data
original_word_counts = X_original.sum(axis=0)
cleaned_word_counts = X_cleaned.sum(axis=0)

# Step 6: Create DataFrames for Visualization
# Convert word counts into DataFrames, showing the top 10 most frequent words
original_feature_counts = pd.DataFrame(
    original_word_counts.A1, 
    index=vectorizer_original.get_feature_names_out(), 
    columns=['Count']
).sort_values(by='Count', ascending=False).head(10)

cleaned_feature_counts = pd.DataFrame(
    cleaned_word_counts.A1, 
    index=vectorizer_cleaned.get_feature_names_out(), 
    columns=['Count']
).sort_values(by='Count', ascending=False).head(10)

# Step 7: Visualization of Word Frequencies Before and After Cleaning
plt.figure(figsize=(14, 6))

# Plot for original reviews
plt.subplot(1, 2, 1)
original_feature_counts.plot(kind='bar', legend=False, color='skyblue')
plt.title('Top 10 Words Before Cleaning')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

# Plot for cleaned reviews
plt.subplot(1, 2, 2)
cleaned_feature_counts.plot(kind='bar', legend=False, color='lightgreen')
plt.title('Top 10 Words After Cleaning')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
